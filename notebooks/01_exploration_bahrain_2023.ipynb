```python
import pandas as pd
import matplotlib.pyplot as plt

# Notebook-Konfiguration für schönere Plots
plt.style.use("seaborn-v0_8")
plt.rcParams["figure.figsize"] = (12, 5)

# Clean Dataset laden
df = pd.read_csv("../data/processed/f1_2023_round1_R_clean.csv")

# Vorschau
df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.hist(df["LapTime_s"], bins=50, color="royalblue", alpha=0.8)
plt.title("Verteilung der Rundenzeiten (in Sekunden)")
plt.xlabel("Laptime [s]")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
plt.scatter(df["TyreLife"], df["LapTime_s"], s=12, alpha=0.5)
plt.title("Reifenalter vs. Rundenzeit")
plt.xlabel("TyreLife (Runden)")
plt.ylabel("LapTime [s]")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df.boxplot(column="LapTime_s", by="Driver", rot=90)
plt.title("Pace-Vergleich der Fahrer (LapTime_s)")
plt.suptitle("")
plt.xlabel("Driver")
plt.ylabel("LapTime [s]")
plt.show()
```
